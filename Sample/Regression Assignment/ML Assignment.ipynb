```python
dataset.columns
Independant = dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
Dependant = dataset[['charges']]
from sklearn.model_selection  import train_test_split
xtrain, xtest,ytrain,ytest = train_test_split(Independant,Dependant,test_size=0.3,random_state=0)
```

```python
from sklearn.preprocessing  import StandardScaler
sc = StandardScaler()
xtrain = sc.fit_transform(xtrain)
xtest = sc.transform(xtest)
```

```python
AlogrithmResultLst = [[]]
```

```python
from sklearn.linear_model import LinearRegression
regressor  = LinearRegression()
x_predict = regressor.fit(xtrain,ytrain)
#x_predict
y_predict = regressor.predict(xtest)
from sklearn.metrics import r2_score
r2sqr = r2_score(ytest,y_predict)
r2sqr
X_Train = xtrain
Y_Train = ytrain
X_Test = xtest
Y_Test = ytest

#To import SVR class
from sklearn.svm import SVR
import warnings
warnings.filterwarnings('ignore')
#An object SVR  is created and assigned to Regressor
kernel={ 'poly', 'rbf', 'sigmoid'}#, 'precomputed','linear',}
#kernel={ 'linear'}
lst = [['',0.0,0.0]] 
for kitem in kernel:    
    for citem in np.arange(0,3100,100):    
    #for citem in [1,10.500,1000.2000,3000]:
        if(citem==0.0):
            citem = 1.0
        if(citem==200.00):
            continue
        #print(f'citem - kitem: {citem},{kitem}')
        Regressor = SVR(C=citem,kernel=kitem)
        #Creates the AI model here with Train data sets
        Regressor.fit(X_Train,Y_Train)
        Y_Predict = Regressor.predict(X_Test)
        # libarary to calculae prediction metrics 2 score        
        #r_scroe callculated and assigned to a variable
        r_sqr = r2_score(Y_Test,Y_Predict)
        #r_sqr
        lst.append([kitem,citem,r_sqr])
        #print(f'citem - kitem - r_sqr: {citem},{kitem},{r_sqr}')        
df = pd.DataFrame(lst, columns =['kernel', 'citem','r_sqr']) 
del lst[0]
from io import StringIO
from tabulate import tabulate
print(tabulate(df, headers='keys', tablefmt='fancy_grid'))
maxdmodel_r_sqr = df.max()
# Get the rows where 'r_sqr' equals the maximum value
maxval =maxdmodel_r_sqr['r_sqr']
matching_rows = df[df['r_sqr'] == maxval]
print( f'Best fit at {matching_rows["r_sqr"].values} in the run- model combination is criterion:{matching_rows["kernel"].values}, splitter:{matching_rows["citem"].values}.')
AlogrithmResultLst.append([matching_rows["r_sqr"].values,"LinearRegression"])
```

```python
#To import LeearRegression class
from sklearn  import tree
from sklearn.tree  import DecisionTreeRegressor 
from sklearn.metrics import r2_score
import matplotlib.pylab as plt

#import random as rd;
#An object DecisionTree  is created and assigned to Regressor
lst = [[]]
criterions = {'squared_error', 'friedman_mse', 'absolute_error', 'poisson'}
splitters = {'best', 'random'}
max_features = {'sqrt', 'log2'}
#rvalues = rd.random()
for critrn in criterions:    
    for split in splitters:
        for feature in max_features:                                     
            Regressor = DecisionTreeRegressor(criterion= critrn,splitter=split,max_features=feature)            
            Regressor.fit(X_Train,Y_Train)
            #tree.plot_tree(Regressor)            
            Y_Predict = Regressor.predict(X_Test)            
            r_sqr = r2_score(Y_Test,Y_Predict)            
            lst.append([critrn,split,feature,r_sqr])            
del lst[0]
df = pd.DataFrame(lst, columns =['criterion', 'splitter','max_features','r_sqr']) 
from io import StringIO
from tabulate import tabulate
print(tabulate(df, headers='keys', tablefmt='fancy_grid'))
maxdmodel_r_sqr = df.max()
maxval =maxdmodel_r_sqr['r_sqr']
matching_rows = df[df['r_sqr'] == maxval]
print( f'Best fit at {matching_rows["r_sqr"].values} in the run- model combination is criterion:{matching_rows["criterion"].values}, splitter:{matching_rows["splitter"].values},max_features:{matching_rows["max_features"].values}' )     
AlogrithmResultLst.append([matching_rows["r_sqr"].values,"DecisionTreeRegressor"])
```

```python
from sklearn.ensemble import RandomForestRegressor
"""regressor = RandomForestRegressor(n_estimators=50,random_state=0)
print (X_Train)
print (Y_Train)
regressor
regressor.fit(X_Train,Y_Train)
_predict = regressor.predict(X_Test)
Y_predict = regressor.predict(X_Test)
rsqr = r2_score(Y_Test,Y_predict)"""

lst = [[]]
criterions = {'squared_error', 'friedman_mse', 'absolute_error', 'poisson'}
splitters = {'best', 'random'}
max_features = {'sqrt', 'log2',None}
for critrn in criterions:        
    for feature in max_features: 
        for estimators in np.arange(50,150,10):            
            Regressor = RandomForestRegressor(criterion= critrn,max_features=feature,n_estimators=estimators)        
            Regressor.fit(X_Train,Y_Train)           
            Y_Predict = Regressor.predict(X_Test)                             
            r_sqr = r2_score(Y_Test,Y_Predict)        
            lst.append([critrn,( feature,'None')[feature==None],r_sqr,estimators])            
del lst[0]
df = pd.DataFrame(lst, columns =['criterion', 'max_features','r_sqr','estimators']) 
from io import StringIO
from tabulate import tabulate
print(tabulate(df, headers='keys', tablefmt='fancy_grid'))
#maxdmodel_r_sqr = df['r_sqr'].max()
maxval =df['r_sqr'].max()
matching_rows = df[df['r_sqr'] == maxval]
print( f'Best fit at {matching_rows["r_sqr"].values} in the run- model combination is criterion:{matching_rows["criterion"].values}, max_features:{matching_rows["max_features"].values}, estimators:{matching_rows["estimators"].values}' )     
AlogrithmResultLst.append([matching_rows["r_sqr"].values,"RandomForestRegressor"])
#del AlogrithmResultLst[0]
AlogrithmResultdf = pd.DataFrame(AlogrithmResultLst, columns =['r_sqr', 'Alogrithm']) 

maxval =AlogrithmResultdf['r_sqr'].max()[0]
matching_rows = AlogrithmResultdf[AlogrithmResultdf['r_sqr'] == maxval]
print( f'Best Algorithm is with {matching_rows["r_sqr"].values[0]} in the model {matching_rows["Alogrithm"].values}' )     
```
