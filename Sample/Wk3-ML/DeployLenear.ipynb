```python
# It is a library used to suppress warnings displayed in the standard output.
import warnings 
# ickle is a library used to save a model generated through Python code into a file with a `.sav` extension.
# The file is stored in binary format, and the saved model can be loaded using the `load` function from the Pickle library. This is mainly used for deployment purposes, allowing the model to be ported and reused in different environments.
import pickle 
#warnings.filterwarnings('error', category=DeprecationWarning)
# The function is used to suppress or ignore warnings generated by the Python interpreter while executing the source code.
warnings.filterwarnings('ignore') 
# The Pickle function `load` is used to load a model from a file located at a specified path (in this case, the working environment folder).
# The loaded model is assigned to the `regressor` variable, which can then be used for further operations, such as making predictions.
# The `LinearRegression` class (or any other class used to create the model) is saved with its initialized values, making the model available for future use.
regressor = pickle.load(open('finalmodel.sav','rb')) 
print("Enter -1 to exit")
while True:
    # Gets the years of experience from the end user to find the salary of the employee from HR.
    YearsOfExperience = int(input("Enter your years of experience to predict your salary"))
    # Checks the condition to end the program if the input is -1 
    if(YearsOfExperience==-1):
        break;
    # The `predict` function in the `LinearRegression` class is used to predict the salary based on the input provided (such as years of experience).
    print("Your salary is ",regressor.predict([[YearsOfExperience]]))
```
